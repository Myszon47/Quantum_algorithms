```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram

def encode_bits(qc, a: str, b: str):
    """Encode binary input values into the quantum registers."""
    a = a[::-1]
    b = b[::-1]
    for i in range(len(a)):
        if a[i] == '1':
            qc.x(i)  # Set `a` qubits
        if b[i] == '1':
            qc.x(i + 4)  # Set `b` qubits
    return qc

def full_adder(qc, a, b, carry_in, sum_out, carry_out):
    """Full adder logic using CCX and CX gates."""
    qc.cx(a, sum_out)
    qc.cx(b, sum_out)
    qc.ccx(a, b, carry_out)
    qc.cx(carry_in, sum_out)
    qc.ccx(carry_in, sum_out, carry_out)

def ripple_carry_adder(qc, qa, qb):
    """Quantum ripple-carry adder using only the allowed gate types."""
    carry_in = QuantumRegister(1, 'carry')
    qc.add_register(carry_in)

    # Apply full adders across all 4-bit positions
    full_adder(qc, qb[0], qa[0], carry_in[0], qa[0], carry_in[0])
    full_adder(qc, qb[1], qa[1], carry_in[0], qa[1], carry_in[0])
    full_adder(qc, qb[2], qa[2], carry_in[0], qa[2], carry_in[0])
    full_adder(qc, qb[3], qa[3], carry_in[0], qa[3], carry_in[0])

# Create the quantum and classical registers
qa = QuantumRegister(4, 'a')
qb = QuantumRegister(4, 'b')
ca = ClassicalRegister(4, 'result')
qc = QuantumCircuit(qa, qb, ca)

# Example input: Binary strings representing two numbers in the range 0-15
a = '0110'  # 6 in decimal
b = '0111'  # 7 in decimal

# Encode the binary inputs into the quantum circuit
qc = encode_bits(qc, a, b)
qc.barrier()

# Apply the ripple-carry adder to add `b` to `a`
ripple_carry_adder(qc, qa, qb)
qc.barrier()

# Measure the resulting value in the first four qubits
qc.measure(qa, ca)

# Simulate the circuit using the QASM simulator
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, backend=simulator, shots=1024).result()

# Plot the resulting distribution of outcomes
plot_histogram(result.get_counts(qc))
```
